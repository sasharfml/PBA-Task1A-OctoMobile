# Import Dependencies

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

```

# Define Corpus

```python
# 1. Dataset contoh kasus banjir bandang
corpus = [
    "Banjir bandang menerjang permukiman warga setelah hujan lebat berkepanjangan dan sungai meluap membawa lumpur.",
    "Tim gabungan melakukan evakuasi warga terdampak banjir ke posko darurat dan menyalurkan bantuan logistik.",
    "Pemerintah daerah mempercepat normalisasi sungai serta perbaikan drainase untuk mengurangi risiko banjir susulan.",
    "A",
    "Upaya evakuasi saat banjir membutuhkan koordinasi cepat antarinstansi",
    "banjir",
    "banjir evakuasi"
]

# 2. Representasi vektor (fokus pada kata kunci utama)
vectorizer = CountVectorizer(vocabulary=['banjir', 'evakuasi'])
X = vectorizer.fit_transform(corpus)
```

```python
# 3. Hitung Cosine Similarity
# Membandingkan semua dokumen terhadap Dokumen 0
cos_sim = cosine_similarity(X[0:1], X)

# Tampilkan hasil
df = pd.DataFrame(X.toarray(), columns=['banjir', 'evakuasi'])
df['Cosine Sim to Doc 0'] = cos_sim.flatten()
print(df)
```

# Cosine Similarity Calculation

**CountVectorizer:** Mengubah teks menjadi angka (matriks frekuensi kata). Pada contoh ini, fitur dibatasi pada kata **banjir** dan **evakuasi** agar fokus pada topik bencana.

**Cosine Similarity vs Distance:** Nilai berada pada rentang 0 sampai 1. Semakin mendekati 1, semakin mirip arah/topik kedua dokumen.

**Analisis Hasil:** Dokumen dengan kombinasi kata *banjir* dan *evakuasi* cenderung memiliki kemiripan lebih tinggi terhadap dokumen utama tentang banjir bandang.

**Euclidean Distance:** Jika hanya memakai jarak lurus, perbedaan panjang dokumen bisa terlihat lebih besar walaupun topiknya mirip.

**Aplikasi Nyata:** Cosine similarity banyak dipakai untuk pencarian berita serupa, deteksi duplikasi konten, dan sistem rekomendasi artikel kebencanaan.

```python
import numpy as np
def length(v):
    return np.sqrt((v ** 2).sum())
def cos_dist(v, w):
    return 1 - (v * w).sum() / (length(v) * length(w))
cos_dist(tf.loc[0], tf.loc[1]), cos_dist(tf.loc[0], tf.loc[2])
```
